```python
import os, time
import torch
import torch.nn.functional as F
import pandas as pd
import torchvision
import torchvision.transforms as transforms

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("DEVICE:", DEVICE)

CKPTS = [
  "/content/drive/MyDrive/final/mnist-variant-topk24-best.pt",
  "/content/drive/MyDrive/final/mnist-variant-topk48-best.pt",
  "/content/drive/MyDrive/final/cifar10-variant-topk24-best.pt",
  "/content/drive/MyDrive/final/cifar10-variant-topk48-best.pt",
  "/content/drive/MyDrive/final/mnist-variant-bw4-best.pt",
  "/content/drive/MyDrive/final/mnist-variant-bw2-best.pt",
  "/content/drive/MyDrive/final/cifar10-variant-bw4-best.pt",
  "/content/drive/MyDrive/final/cifar10-variant-bw2-best.pt",
  "/content/drive/MyDrive/final/mnist-variant-blockdiag2x2.pt",
  "/content/drive/MyDrive/final/mnist-baseline.pt",
  "/content/drive/MyDrive/final/cifar10-variant-blockdiag2x2.pt",
  "/content/drive/MyDrive/final/cifar10-baseline.pt",
]

missing = [p for p in CKPTS if not os.path.exists(p)]
if missing:
    print("Missing files:", missing)
else:
    print("All checkpoint files found ")
```

```python
def make_test_loader(dataset: str, batch_size: int = 256, num_workers: int = 2):
    if dataset == "mnist":
        tfm = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.5,), (0.5,))
        ])
        ds = torchvision.datasets.MNIST("./data", train=False, download=True, transform=tfm)
    elif dataset == "cifar10":
        tfm = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))
        ])
        ds = torchvision.datasets.CIFAR10("./data", train=False, download=True, transform=tfm)
    else:
        raise ValueError(dataset)

    return torch.utils.data.DataLoader(
        ds, batch_size=batch_size, shuffle=False, num_workers=num_workers,
        pin_memory=torch.cuda.is_available()
    )
```

```python
def infer_dataset_from_name(name: str) -> str:
    if name.startswith("/content/drive/MyDrive/final/mnist"):
        return "mnist"
    if name.startswith("/content/drive/MyDrive/final/cifar10"):
        return "cifar10"
    raise ValueError(f"Can't infer dataset from: {name}")

def infer_variant_from_name(name: str) -> str:
    if "baseline" in name:
        return "baseline"
    if "blockdiag2x2" in name:
        return "blockdiag2x2"
    if "topk" in name:
        return "topk"
    if "bw" in name:
        return "banded"
    return "unknown"

def build_model_for_ckpt(ckpt_name: str, config: dict):
    variant = infer_variant_from_name(ckpt_name)

    if variant in ["baseline", "blockdiag2x2"]:
        import main as M
        return M.ViT(config)

    if variant == "banded":
        import banded_variant as M
        return M.ViT(config)

    if variant == "topk":
        import top_k_variant as M
        return M.ViT(config)

    raise ValueError(f"Unknown variant for {ckpt_name}")
```

```python
@torch.no_grad()
def eval_acc_loss(model, loader):
    model.eval()
    total, correct, loss_sum = 0, 0, 0.0
    for x, y in loader:
        x = x.to(DEVICE, non_blocking=True)
        y = y.to(DEVICE, non_blocking=True)
        logits = model(x)
        loss = F.cross_entropy(logits, y)
        bs = x.size(0)
        total += bs
        correct += (logits.argmax(dim=-1) == y).sum().item()
        loss_sum += loss.item() * bs
    return correct / total, loss_sum / total

@torch.no_grad()
def infer_ms_per_batch(model, x, iters=200, warmup=50):
    model.eval()
    x = x.to(DEVICE, non_blocking=True)

    if DEVICE == "cuda":
        starter = torch.cuda.Event(enable_timing=True)
        ender = torch.cuda.Event(enable_timing=True)

        for _ in range(warmup):
            _ = model(x)
        torch.cuda.synchronize()

        times = []
        for _ in range(iters):
            starter.record()
            _ = model(x)
            ender.record()
            torch.cuda.synchronize()
            times.append(starter.elapsed_time(ender))  # ms
        return sum(times) / len(times)

    else:
        for _ in range(warmup):
            _ = model(x)
        t0 = time.perf_counter()
        for _ in range(iters):
            _ = model(x)
        t1 = time.perf_counter()
        return 1000.0 * (t1 - t0) / iters

def pretty_model_label(name: str) -> str:
    if "baseline" in name:
        return "Baseline (dense)"
    if "blockdiag2x2" in name:
        return "BlockDiag2×2"
    if "bw2" in name:
        return "Banded (bw=2)"
    if "bw4" in name:
        return "Banded (bw=4)"
    if "topk24" in name:
        return "TopK (k=24)"
    if "topk48" in name:
        return "TopK (k=48)"
    return name.replace(".pt","")

rows = []
for path in CKPTS:
    ckpt = torch.load(path, map_location="cpu")
    meta = ckpt.get("meta", {})
    config = meta.get("config", None)
    if config is None:
        raise RuntimeError(f"{path}: checkpoint missing meta['config']")

    dataset = meta.get("dataset", infer_dataset_from_name(path))
    model = build_model_for_ckpt(path, config)
    model.load_state_dict(ckpt["model_state_dict"], strict=True)
    model.to(DEVICE)

    # params
    params_m = sum(p.numel() for p in model.parameters()) / 1e6

    # evaluate
    test_loader = make_test_loader(dataset, batch_size=256)
    acc, loss = eval_acc_loss(model, test_loader)

    # inference timing on fixed batch size
    xb, _ = next(iter(test_loader))
    xb = xb[:128]
    ms_batch = infer_ms_per_batch(model, xb, iters=200, warmup=50)
    ms_img = ms_batch / xb.size(0)

    # training time
    epoch_times = ckpt.get("epoch_times", [])
    train_s_epoch = (sum(epoch_times) / len(epoch_times)) if epoch_times else None
    train_s_total = sum(epoch_times) if epoch_times else None

    rows.append({
        "Dataset": dataset.upper(),
        "Model": pretty_model_label(path),
        "Params (M)": round(params_m, 3),
        "Best Acc (ckpt)": float(ckpt.get("best_acc", acc)),
        "Test Acc (now)": acc,
        "Train s/epoch": train_s_epoch,
        "Train s_total": train_s_total,
        "ms/batch (128)": ms_batch,
        "ms/img": ms_img,
        "ckpt": path
    })

df = pd.DataFrame(rows).sort_values(["Dataset", "Model"])
df
```

```python
import matplotlib.pyplot as plt

def get_hist(ckpt):
    return (
        ckpt.get("train_losses", []),
        ckpt.get("test_losses", []),
        ckpt.get("accuracies", []),
        ckpt.get("epoch_times", []),
    )

# group by dataset
by_ds = {"MNIST": [], "CIFAR10": []}
for path in CKPTS:
    ds = infer_dataset_from_name(path)
    ds_key = "MNIST" if ds == "mnist" else "CIFAR10"
    ckpt = torch.load(path, map_location="cpu")
    tr, te, acc, et = get_hist(ckpt)
    by_ds[ds_key].append((pretty_model_label(path), tr, te, acc, et))

fig, axes = plt.subplots(2, 4, figsize=(20, 7), dpi=150)
cols = ["Train Loss", "Test Loss", "Accuracy", "Epoch Time (s)"]

for r, ds_key in enumerate(["MNIST", "CIFAR10"]):
    for c in range(4):
        axes[r, c].set_title(f"{ds_key}: {cols[c]}")
        axes[r, c].set_xlabel("Epoch")
        axes[r, c].grid(True, alpha=0.3)

    for label, tr, te, acc, et in by_ds[ds_key]:
        if tr:
            axes[r,0].plot(range(1, len(tr)+1), tr, marker="o", linewidth=1, label=label)
        if te:
            axes[r,1].plot(range(1, len(te)+1), te, marker="o", linewidth=1, label=label)
        if acc:
            axes[r,2].plot(range(1, len(acc)+1), acc, marker="o", linewidth=1, label=label)
        if et:
            axes[r,3].plot(range(1, len(et)+1), et, marker="o", linewidth=1, label=label)


    for c in [0,1,2]:
        axes[r,c].legend(fontsize=8)

plt.tight_layout()
out_path = "/content/drive/MyDrive/final/training_curves_all_variants.png"
plt.savefig(out_path)
print("Saved:", out_path)
```
